Cek nilai duplikat dan nilai unik

Nilai duplikat: nilai yang muncul lebih dari satu kali dalam dataframe.
Nilai Unik: nilai yang hanya muncul sekali dalam datafarme.

```python
important_columns = ['NAMA PASIEN', 'Usia', 'Jenis_Kelamin', 'Durasi_Sesi']

duplikat = data[data.duplicated(subset=important_columns, keep=False)]

print("Nilai Duplikat:")
print(duplikat)
```

```python
unik = data.drop_duplicates(subset=important_columns, keep=False)

print("Nilai Unik:")
print(unik)
```

Visualisasikan jumlah nilai unik

```python
import matplotlib.pyplot as plt

unique_values = data.nunique()

plt.figure(figsize=(10, 6))
unique_values.plot(kind='bar', color='violet')

plt.title('Jumlah Nilai Unik per Kolom')
plt.xlabel('Kolom')
plt.ylabel('Jumlah Nilai Unik')

plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
